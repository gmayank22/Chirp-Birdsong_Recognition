```python
#DATA CLEANING
```

```python
df.GPA=df.GPA.astype(str)
i=["nan","Personal","Unknown"]
for s in i:
  df.GPA=df.GPA.apply(lambda x:x.replace(s,"3.3"))
df.GPA=df.GPA.apply(lambda x:x.split(' ')[0])
df.GPA=df.GPA.astype(float)   #converting GPA to its correct datatype!!
df.GPA.unique()
```

```python
#below, we replace null values in some column with their corresponding median as it gives us an idea that people would more incline towards the meadian and hence it will produce less distortion of data!!
# while in others , null values are replaced with -1 to show negativity of opinion or the person opted out none of the above option during the survey!!
```

```python
#now lets consider what each column represent:
#calories_chicken and calories_scone provides the calories gained by the person intaking chicken and scone respectively!!
```

```python
df.calories_day=df.calories_day.fillna(df.calories_day.median())
df.calories_day=df.calories_day.astype(int)
df.calories_day.unique()
```

```python
#here,calories_day represent avg calorie intake by the person as a whole in a day!!
#since, data here is coded so lets consider its data as follows:
#2 means 2000-2500, 3 means 2500-3000, 4 means 3000-3500
```

```python
df.cook=df.cook.fillna(df.cook.median())
df.cook=df.cook.astype(int)
df.cook.unique()
```

```python
#here,cook represent who might cook food for the person, so,lets assume our data as follows:
#1 means they cook themselves,2 means their father cook for them, 3 means their mother cook for them, 4 means their siblings cook for them, 5 means their servants cook for them!!
```
